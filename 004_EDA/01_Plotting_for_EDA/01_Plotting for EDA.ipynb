# Histogram PDF, CDF

```python
# Need for Cumulative Distribution Function (CDF)
# We can visually see what percentage of versicolor flowers have a
# petal length of less than 1.6?
# How to construct a CDF?
# How to read a CDF?

# Plot CDF of petal length

counts, bin_edges=np.histogram(iris_setosa['PetalLengthCm'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

# Compute CDF.
cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
# Plots of CDF of petal length for various types of flowers.
# Misclassification error if you use petal length only.

#Plot CDF of petal_ length

# PDF and CDF for the Setosa.
counts, bin_edges=np.histogram(iris_setosa['PetalLengthCm'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

# Compute CDF.
cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

# PDF and CDF for the Verginica.
counts, bin_edges=np.histogram(iris_virginica['PetalLengthCm'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

# Compute CDF.
cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

# PDF and CDF for the Versicolor.
counts, bin_edges=np.histogram(iris_versicolor['PetalLengthCm'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

# Compute CDF.
cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

# Mean, Variance and Standard Deviation

```python
# Mean, Variance and Standard Deviation.

# Mean:

print('Means:')
print(np.mean(iris_setosa['PetalLengthCm']))
print(np.mean(iris_virginica['PetalLengthCm']))
print(np.mean(iris_versicolor['PetalLengthCm']),'\n')

# Mean with an outlier.
print('Mean with an outlier:')
print(np.mean(np.append(iris_setosa['PetalLengthCm'],50)),'\n')

# Standard deviation:

print('Standard Deviation:')
print(np.std(iris_setosa['PetalLengthCm']))
print(np.std(iris_virginica['PetalLengthCm']))
print(np.std(iris_versicolor['PetalLengthCm']))
```

# Median, Percentile, Quantile, IQR, MAD
### $\bigstar{}$ MAD : Median Absolute Deviation
Median absolute deviation are same thing like standard deviation.<br>
The difference is standard deviation is taken around mean and median absolute deviation is taken about median.
### $\bigstar{}$ IQR : Inter Quartile Range
IQR(Inter Quartile Range) is the difference between the $75^{th}$ percentile and $25^{th}$ percentile.<br>


# Percentile and Quantile

$50^{th}$ percentile is nothing but the median.<br>
$x$ percentile means there are $x$ percent values that are less than $x$.<br>
and $(100-x)$ percent values are greater than $x$.

```python
# Median, Percentile, Quantile, IQR, MAD

print('Median:')
print(np.median(iris_setosa['PetalLengthCm']))
print(np.median(iris_virginica['PetalLengthCm']))
print(np.median(iris_versicolor['PetalLengthCm']))

# Median with an outlier

print('\nMedian with an outlier:')
print(np.median(np.append(iris_setosa['PetalLengthCm'], 50)))

# Quantiles

print('\nQuantiles:')
print(np.percentile(iris_setosa['PetalLengthCm'], np.arange(0,100,25)))
print(np.percentile(iris_virginica['PetalLengthCm'], np.arange(0,100,25)))
print(np.percentile(iris_versicolor['PetalLengthCm'], np.arange(0,100,25)))

# 90^th Percentile

print('\n90th Percentile:')
print(np.percentile(iris_setosa['PetalLengthCm'],90))
print(np.percentile(iris_virginica['PetalLengthCm'],90))
print(np.percentile(iris_versicolor['PetalLengthCm'],90))

# Median Absolute Deviation:

from statsmodels import robust
print('\nAbsolute Median Deviation:')
print(robust.mad(iris_setosa['PetalLengthCm']))
print(robust.mad(iris_virginica['PetalLengthCm']))
print(robust.mad(iris_versicolor['PetalLengthCm']))

# NOTE: Median and Median Absolute Deviation will not be corrupted by outliers.
```

# Univariate, Bivariate and Multivariate Analysis.

$\textbf{Univariate Analysis:}$ Univariate analysis focuses on analyzing a single variable in isolation. It involves examining the distribution, summary statistics, and visualizations of a single variable to understand its characteristics. Common techniques used in univariate analysis include histograms, box plots, bar charts, and measures such as mean, median, and standard deviation. <br>

$\textbf{Bivariate Analysis:}$ Bivariate analysis involves the analysis of two variables simultaneously to explore relationships or associations between them. It helps in understanding how changes in one variable affect the other. Bivariate analysis techniques include scatter plots, correlation analysis, contingency tables, and hypothesis testing for assessing the statistical significance of the relationship. <br>

$\textbf{Multivariate Analysis:}$ Multivariate analysis deals with the simultaneous analysis of three or more variables to uncover complex relationships and patterns among them. It explores how multiple variables interact and influence each other. Multivariate analysis techniques include dimensionality reduction techniques (e.g., principal component analysis), cluster analysis, factor analysis, and regression analysis with multiple predictors.
